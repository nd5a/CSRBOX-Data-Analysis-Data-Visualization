Write   a   python   program   to   perform   multiclass   classification   oniris dataset.

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# load data set
iris = load_iris()
x, y = iris.data, iris.target
```

```python
x
```

```python
y
```

```python
# Split data set into training and testing data set
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=50, stratify=y,test_size=0.8)
```

```python
x_train
x_test
y_train
y_test
```

```python
# create and train the KNN classifier

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_test, y_test)
```

```python
# Make predictions

y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=iris.target_names)

print(f'Accuracy: {accuracy:.2f}')
print()
print('Classification Report:\n', report)
```
