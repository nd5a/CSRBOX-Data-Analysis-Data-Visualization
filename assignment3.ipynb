```python
# Visualize the data with graphs and description to each.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data
df = pd.read_csv('cars.csv', encoding='ISO-8859-1')

# Display the first few rows of the dataset
print(df.head())

# Basic statistics
print("\nBasic statistics:")
print(df.describe())

# Check for missing data
print("\nMissing data check:")
print(df.isnull().sum())

# Plot 1: Distribution of Miles per Gallon (mpg)
plt.figure(figsize=(10,6))
sns.histplot(df['mpg'], kde=True, color='blue', bins=20)
plt.title('Distribution of Miles per Gallon (mpg)')
plt.xlabel('Miles per Gallon (mpg)')
plt.ylabel('Frequency')
plt.show()

# Plot 2: Car Weight vs Miles per Gallon (mpg)
plt.figure(figsize=(10,6))
sns.scatterplot(x='wt', y='mpg', data=df, color='green')
plt.title('Car Weight vs Miles per Gallon (mpg)')
plt.xlabel('Weight (in 1000 lbs)')
plt.ylabel('Miles per Gallon (mpg)')
plt.show()

# Plot 3: Horsepower vs Miles per Gallon (mpg)
plt.figure(figsize=(10,6))
sns.scatterplot(x='hp', y='mpg', data=df, color='red')
plt.title('Horsepower vs Miles per Gallon (mpg)')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Miles per Gallon (mpg)')
plt.show()
```
